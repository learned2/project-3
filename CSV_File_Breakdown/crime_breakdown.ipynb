```python
#import dependencies
import pandas as pd
import numpy as np
```

```python
#Bring in CSV
crime_csv = pd.read_csv('../../NYPD_Complaint_Data_Current__Year_To_Date__20250320.csv');
```

```python
#Create DF with csv
crime_data_frame = pd.DataFrame(crime_csv)
#Show all columns
pd.set_option('display.max_columns', None)
#Display DF
crime_data_frame.head()
```

```python
#Create new df with just the important columns
consolidated_crime_df = crime_data_frame.copy()
consolidated_crime_df = consolidated_crime_df[['CMPLNT_NUM', 'CMPLNT_FR_DT', 'CMPLNT_FR_TM', 'OFNS_DESC',
                                               'CRM_ATPT_CPTD_CD', 'BORO_NM', 
                                               'Latitude', 'Longitude']]
#Filter for the crimes we are looking into
offenses = ['MURDER & NON-NEGL. MANSLAUGHTER', 'ARSON', 'GRAND LARCENY OF MOTOR VEHICLE']
consolidated_crime_df = consolidated_crime_df[consolidated_crime_df['OFNS_DESC'].isin(offenses)]
#Display the data frame
consolidated_crime_df.head()
                                               
```

```python
#Display the unique values in the OFNS_DESC (Offense Description) column to see which values we would like to isolate
consolidated_crime_df['OFNS_DESC'].unique()
```

```python
def csv_save(df, crime_type: str):
    crime_csv_filename = f"nyc_{crime_type}.csv"
    df.to_csv(crime_csv_filename, index = False)
```

```python
def crime_df_creator(df, crime_type: str):
    crime_df = df.copy()
    crime_df = crime_df[['CMPLNT_NUM', 'CMPLNT_FR_DT', 'CMPLNT_FR_TM', 'OFNS_DESC', 'Latitude', 'Longitude']]
    crime_df = crime_df[crime_df['OFNS_DESC'] == crime_type]
    return crime_df
```

```python
#Create DataFrame for just Murder
murder_df = crime_df_creator(consolidated_crime_df, 'MURDER & NON-NEGL. MANSLAUGHTER')

#Create DataFrame for just Arson
arson_df = crime_df_creator(consolidated_crime_df, 'ARSON')

#Create DataFrame for just Grand Larceny of Motor Vehicle
gta_df = crime_df_creator(consolidated_crime_df, 'GRAND LARCENY OF MOTOR VEHICLE')
```

```python
#Save the file as a csv
murder_csv = csv_save(murder_df, 'murder')

#Save the file as a csv
arson_csv = csv_save(arson_df, 'arson')

#Save the file as a csv
gta_csv = csv_save(gta_df, 'larceny_vehicles')
```
